```python
# Simulating a simple board game with two players, 
# each player taking a turn to roll a die and moving at that number of spaces. 
# The first person to space 100 wins

import random
class Player:
    def __init__(self, name):
        self.name = name
        self.position = 0 ##The starting position

    def roll_die(self): ##Simulating rolling a die
        return random.randint(1,6)
    
    def move(self):
        roll = self.roll_die()
        self.position += roll
        print(f"{self.name} rolled a {roll} and moved to space {self.position}")
class Game:
    def __init__(self, spaces, player1, player2):
        self.spaces = spaces
        self.player1 = player1
        self.player2 = player2

    def start(self): ##Running the loop until one player reaches the end
        print("Game Starting..\n")
        while self.player1.position < self.spaces and self.player2.position < self.spaces:
            self.player1.move() ##Player 1's turn
            if self.player1.position >= self.spaces:
                print(f"\n {self.player1.name} wins the game!")
                break
            ##Player 2's turn
            self.player2.move()
            if self.player2.position >= self.spaces:
                print(f"\n {self.player2.name} wins the game")
                break

        ##Now Creating the two players
p1 = Player("Elijah")
p2 = Player("Elisha")

        ##Creating and starting the game with 100 spaces
game = Game(100, p1, p2)
game.start()

    
        
        
```
